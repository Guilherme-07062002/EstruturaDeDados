```python
#Gerando gráficos com os dados obtidos utilizando python

A primeira celula de código abaixo, constroi os gráficos com base no objeto *data*.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = {
    'Algoritmo': ['Bubble Sort', 'Selection Sort', 'Insertion Sort', 'Shell Sort', 'Quick Sort', 'Merge Sort'],
    '1,000(Mil) elementos': [13.1, 5.29, 2.71, 0.257, 0.90, 0.79],
    '10,000(Dez mil) elementos': [873.2, 262.9, 175.0, 2.55, 0.16, 2.8],
    '100,000(Cem mil) elementos': [851.44, 307.99, 277.44, 29.3, 6.888, 1.58]
}

df = pd.DataFrame(data)

# Gráfico para 1,000(Mil) elementos
df.plot(kind='line', x='Algoritmo', y='1,000(Mil) elementos',
        title='Tempo de Execução dos Algoritmos - 1,000(Mil) elementos')
plt.ylabel('Tempo (ms)')

# Gráfico para 10,000(Dez mil) elementos
df.plot(kind='line', x='Algoritmo', y='10,000(Dez mil) elementos',
        title='Tempo de Execução dos Algoritmos - 10,000(Dez mil) elementos')
plt.ylabel('Tempo (ms)')

# Gráfico para 100,000(Cem mil) elementos
df.plot(kind='line', x='Algoritmo', y='100,000(Cem mil) elementos',
        title='Tempo de Execução dos Algoritmos - 100,000(Cem mil) elementos')
plt.ylabel('Tempo (ms)')

plt.show()
```

Abaixo, este código gera os gráficos com base no arquivo tempos_execucao.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Ler o arquivo CSV
df = pd.read_csv('./algoritmosDeOrdenação/tempos_execucao.csv')

# Gráfico para 100 elementos
df_100 = df[df['Tamanho do Array'] == 100]
df_100.plot(kind='line', x='Algoritmo', y='Tempo de Execucao (ms)',
            title='Tempo de Execução dos Algoritmos - 100 elementos')
plt.ylabel('Tempo (ms)')

# Gráfico para 1.000 elementos
df_1000 = df[df['Tamanho do Array'] == 1000]
df_1000.plot(kind='line', x='Algoritmo', y='Tempo de Execucao (ms)',
             title='Tempo de Execução dos Algoritmos - 1.000 elementos')
plt.ylabel('Tempo (ms)')

# Gráfico para 10.000 elementos
df_10000 = df[df['Tamanho do Array'] == 10000]
df_10000.plot(kind='line', x='Algoritmo', y='Tempo de Execucao (ms)',
              title='Tempo de Execução dos Algoritmos - 10.000 elementos')
plt.ylabel('Tempo (ms)')

plt.show()
```
