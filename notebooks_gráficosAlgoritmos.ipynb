```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = {
    'Algoritmo': ['Bubble Sort', 'Selection Sort', 'Insertion Sort', 'Shell Sort', 'Quick Sort', 'Merge Sort'],
    '1,000(Mil) elementos': [10.801, 3.382, 1.625, 0.800, 2.292, 0.461],
    '10,000(Dez mil) elementos': [733.049, 226.285, 146.256, 64.380, 28.781, 4.725],
    '100,000(Cem mil) elementos': [534.219, 222.929, 180.982, 78.750, 72.587, 62.526]
}

df = pd.DataFrame(data)

# Grfico para 1,000(Mil) elementos
df.plot(kind='bar', x='Algoritmo', y='1,000(Mil) elementos',
        title='Tempo de Execuo dos Algoritmos - 1,000(Mil) elementos')
plt.ylabel('Tempo (ms)')

# Grfico para 10,000(Dez mil) elementos
df.plot(kind='bar', x='Algoritmo', y='10,000(Dez mil) elementos',
        title='Tempo de Execuo dos Algoritmos - 10,000(Dez mil) elementos')
plt.ylabel('Tempo (ms)')

# Grfico para 100,000(Cem mil) elementos
df.plot(kind='bar', x='Algoritmo', y='100,000(Cem mil) elementos',
        title='Tempo de Execuo dos Algoritmos - 100,000(Cem mil) elementos')
plt.ylabel('Tempo (ms)')

plt.show()
```
